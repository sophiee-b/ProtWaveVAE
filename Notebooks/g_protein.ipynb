```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
df = pd.read_csv("Pfam_analysis/outputs/prediction/pfam/GProtein/Gprotein_train_dataset_inference.csv")
z_columns = ['z_0', 'z_1', 'z_2', 'z_3', 'z_4', 'z_5']
df['phylogeny_group'] = df['phylogeny'].apply(lambda x: x.split(',')[1] if ',' in x else x)
df_subset1 = df[df['phylogeny_group'].isin(["Metazoa","Fungi","Amoebozoa","Viridiplantae","Alveolata"])]
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_subset1[z_columns])
df_subset1['PCA1'] = pca_result[:, 0]
df_subset1['PCA2'] = pca_result[:, 1]

plt.figure(figsize=(10, 8))
sns.scatterplot(x='PCA1', y='PCA2', hue='phylogeny_group', data=df_subset1, palette='tab10', s=100, alpha=0.7)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Phylogeny Group", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
Please use pip<24.1 if you need to use this version.df_subset2 = df[df['phylogeny_group'].isin(["Proteobacteria","Metazoa","Chlamydiae","Fungi","Actinobacteria"])]

pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_subset2[z_columns])
df_subset2['PCA1'] = pca_result[:, 0]
df_subset2['PCA2'] = pca_result[:, 1]

plt.figure(figsize=(10, 8))
sns.scatterplot(x='PCA1', y='PCA2', hue='phylogeny_group', data=df_subset2, palette='tab10', s=100, alpha=0.7)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Phylogeny Group", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
